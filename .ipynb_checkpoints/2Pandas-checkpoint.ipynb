# Manipulating Data

#Thing to remember: "Drop Function Takes a label NOT POSITION"

```python
df.drop(0)      #Meaning: It removes the row which has label=0
df.head()
```

Drop doesn't remove the row in place, it makes a copy without including 0th row

```python
a = df.drop(0)
a.head()
```

To drop the row in df itself, we have to pass another argument i.e. inplace=True

```python
df.drop(0,inplace=True)
df.head()
```

When inplace = True , the data is modified in place, which means it will return nothing and the dataframe is now updated. When inplace = False , which is the default, then the operation is performed and it returns a copy of the object. You then need to save it to something.

```python
df.drop(0,inplace=True) #when we run the above code again, it we throw an error 
df.head()
#KeyError: '[0] not found in axis'
```

```python
df.drop(3,inplace=True)
df.head()
```

```python
df.index             #way to look at labels
```

### Handling NaN entries

We will do 2 things:
    1. we will drop the rows which are having the NaN
    2. we will replace the NaN with something else

```python
df.iloc[2:4,1:3]=0
df.head()
```

```python
df.iloc[2:4,1:3] = np.nan #nan is a constant from numpy(NaN=Not a Number)
df.head()
```

```python
df.describe()
```

we have inbuilt dropna() function to drop all the rows having na entries in it

#### Droping the NaN 

```python
df.dropna(inplace=True)
df.head()
```

```python
df.reset_index(drop=True, inplace=True)
df.head()
```

#### Replacing the NaN

```python
df.iloc[2:4,1:3] = np.nan 
df.head()
```

We are going to replace the NaN values in the sw column with the mean value of sw column with the help of fillna() function

```python
df.sw.fillna(df.sw.mean(), inplace=True)
```

```python
df.pl.fillna(df.pl.mean(), inplace=True)
```

## String Based Data

```python
df["Gender"] = "Female"
df.iloc[0:10,6] = "Male"
print(df.head())
df.Gender
```

```python
def f(s):
    if s=="Male":
        return 0
    else:
        return 1
    

df["sex"] = df.Gender.apply(f)  #whatever is returned the f function will be stored in the sex column
df.head()
```

```python
del df["Gender"]
df.head()
```
