```python
import matplotlib.pyplot as plt
import numpy as np
```

### Plotting Points

```python
x = [1,2,3]
y = [2,4,6]

plt.scatter(x,y)
plt.show()
```

```python
x = [i for i in range(1,4)]
y = [i for i in range(2,8,2)]

plt.scatter(x,y)
plt.show()
```

### Connecting points with a line

```python
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y)
plt.show()
```

```python
#Here we will get the points and the line as well
x = [1,2,3]
y = [2,4,6]

plt.scatter(x,y)
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3]
y = [2,4,6]

x2 = [3,4,5]
y2 = [2,4,6]
plt.plot(x,y,'r-.')
plt.plot(x2,y2,'y--')
plt.show()
'''Here the first letter of the third argument picks up the color and 
the second argument onwards it picks the type of the points in which
it will connect the line'''
```

```python
#plotting grapg for y = x^3

x = np.array([1,2,3,4])
y = x**3
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,5,0.1)
y = x**3
plt.plot(x,y)
plt.show()
```

If we give only one array and we try to plot the graph, by default for n points it will take "range(n)" i.e.(0 to n-1)

```python
a = [3,4,5]
plt.plot(a)

plt.show()
#it will take a as Y and for x it will take range(n) i.e. X=[0,1,2] 
```

```python
'''
Color Options 
b: blue
g: green
r: red
c: cyan
m: magenta
y: yellow
k: black
w: white
'''
```

## Customizing Graphs

```python
x = np.arange(0,5,0.1)
y = x**3
plt.plot(x,y,color="magenta",marker="+")
plt.show()
```

```python
#color, marker, linewidth, labels for axis and dataset,grid,changing axis, adding text
x = np.arange(0,5,0.1)
y = x**3
y2 = x**2

plt.plot(x,y2,color="red",linewidth=2.5,label="X^2")

plt.plot(x,y,color="cyan",linewidth=1.5,label="X^3")
plt.axis([0,5,0,120])  #for changing the axis range (first two arguments for x and second two for y)

plt.ylabel("Y")
plt.xlabel('X')

plt.legend()             #for displaying the labels of the lines

plt.grid()

plt.text(2,80,"Graph",fontsize=15)          #Three arguments: x,y,text

plt.title("MatplotLib Demo")

plt.show()
```

## Fancy Graphs

```python
sizes = [3,4,6,2]
labels = ["A","B","C","D"]
colors = ["orange","blue","red","green"]
explode = [0.1,0,0,0]
plt.pie(sizes,colors=colors,labels=labels,explode=explode,autopct="%.1f%%")   
plt.title("Split Among Titles")
plt.axis("equal")   #to make the pie symmetrical
plt.show()
```

```python
#autopct for percentage, .1f means it will show one value after decimal
#and %% for showing percentage symbol
#by default the pie graph is counterclockwise, to make it clockwise we have to add counterclock=False
```

```python
marks = [90, 85, 95, 75, 100]
subjects = ['c++', 'java', 'dsa', 'ml', 'webD']
plt.bar(subjects,marks)
```

```python
#histogram
#histograms are useful to visualize how the data is distributed in some buckets of particular range
from random import sample
data = sample(range(1, 1000), 100) #here, we are sampling 100 values between 1 and 1000
#let's say we want to see how many values are sampled between 0-100 or 100-200 (buckets) and so on
#we can use histogram for this
plt.hist(data) #by default, it will plot the data distribution in 10 buckets and on y axis, we
#have the number of values in a particular bucket
```

```python
#boxplot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
```
